```python
import difflib
```

```python

with open('stardust_filters.txt', 'r') as f:
    stardust_filters = [[item for item in line.strip('\n').split(' ') if item] for line in f.readlines()]
with open('eazy_filters.txt', 'r') as f:
    eazy_filters = [[item for item in line.strip('\n').split(' ') if item] for line in f.readlines()]
print(stardust_filters[:3])
```

```python

stardust_names = [item[2] for item in stardust_filters]
stardust_codes = [item[0] for item in stardust_filters]
eazy_names = [item[2] for item in eazy_filters]
eazy_codes = [item[0] for item in eazy_filters]
print(stardust_names[:3])
print(eazy_names[:3])
```

```python
stardust_code_translation = {name:filter_code for name, filter_code in zip(stardust_names,stardust_codes)}
eazy_code_translation = {name:filter_code for name, filter_code in zip(eazy_names,eazy_codes)}
```

```python
stardust_eazy_translation = {}
for stardust_name in stardust_names:
    eazy_match = difflib.get_close_matches(stardust_name, eazy_names, n=1)[0]
    print(f'{stardust_name} -> {eazy_match}')
    stardust_eazy_translation[stardust_code_translation[stardust_name]] = eazy_code_translation[eazy_match]
```

```python
with open('stardust-eazy-filter-translation.txt','w') as f:
    f.write('Stardust Eazy')
    for stardust_code in stardust_eazy_translation:
        eazy_code = stardust_eazy_translation[stardust_code]
        f.write(f'\n{stardust_code} {eazy_code}')
```
